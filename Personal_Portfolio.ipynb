#### 1. เครือข่ายที่เล็กที่สุด (Smallest Network)
เครือข่ายที่เล็กที่สุดคือ เครือข่ายแบบจุดต่อจุด (Point-to-Point) ที่ประกอบด้วยคอมพิวเตอร์เพียงสองเครื่องเชื่อมต่อกันโดยตรงด้วยสายเคเบิล โดยไม่ผ่านอุปกรณ์สวิตช์หรือเราเตอร์ เช่น PC0 และ PC1 ซึ่งมีเพียง 2 โฮสต์ที่เชื่อมต่อกันโดยตรงในเครือข่ายเดียวกัน

#### 2. เครือข่ายที่ต้องใช้สวิตช์ (Network that requires switches)
เครือข่ายที่ต้องใช้สวิตช์คือเครือข่ายที่มีคอมพิวเตอร์หลายเครื่องอยู่ในเครือข่ายเดียวกัน (LAN เดียวกัน) และต้องการให้อุปกรณ์ทั้งหมดสามารถสื่อสารกันผ่านตัวกลางร่วมกัน โดยสวิตช์ทำหน้าที่เชื่อมต่อและส่งต่อข้อมูลภายใน LAN จากตัวอย่างจะเห็นว่า PC2, PC3 และ PC4 ถูกเชื่อมเข้ากับ Switch0 ซึ่งกระจายข้อมูลให้ทุกอุปกรณ์ที่อยู่ใน LAN เดียวกัน

#### 3. เครือข่ายที่ต้องใช้เราเตอร์ (Network that requires routers)
เครือข่ายที่ต้องใช้เราเตอร์คือเครือข่ายที่ต้องเชื่อมต่อหลายเครือข่ายเข้าด้วยกัน เช่น หลาย LAN หรือหลาย Subnet เพราะเราเตอร์ทำหน้าที่เลือกเส้นทางและส่งข้อมูลจากเครือข่ายหนึ่งไปยังอีกเครือข่ายหนึ่ง ในตัวอย่างจะเห็นว่าแต่ละสวิตช์มีกลุ่มโฮสต์ของตนเอง ซึ่งถือเป็นเครือข่ายย่อยคนละเครือข่าย และ Router1 ทำหน้าที่เป็นตัวกลางเชื่อม LAN/Subnet เหล่านี้เข้าด้วยกัน ทำให้สามารถสื่อสารข้ามเครือข่ายได้อย่างถูกต้อง

## **Final Project**
ชื่อโปรเจค **Spacetime Networking — Spacecraft-to-Satellite Communication Network (TS-Com)**

Link(Github, Video, NotebookLM) : [Click here](https://drive.google.com/drive/folders/1pMTUDmsCbXrMSmXztLz72NXbjaL6pmo5?usp=sharing)

รับบทบาทเป็น Engineer มีหน้าที่ในการพัฒนาโปรโตคอลและเขียนโค้ดระบบจำลองทั้งหมด โดยได้พัฒนาระบบครบทุกส่วนตั้งแต่ยานอวกาศ ดาวเทียม ไปจนถึงสถานีภาคพื้นดิน

- เขียนโมดูล sensor.py เพื่อจำลองการทำงานของยานอวกาศที่ส่งข้อมูล Telemetry ผ่าน UDP พร้อมสถานะต่างๆ เช่น สถานะปกติ เตือนภัย หรือฉุกเฉิน
- พัฒนา client.py สำหรับดาวเทียม(relay) ซึ่งมีกลไกสำคัญคือ Mission QoS ที่จัดลำดับความสำคัญของข้อมูลตามภารกิจ โดยข้อมูลฉุกเฉิน(P0) จะถูกส่งก่อนข้อมูลทั่วไป(P2) เสมอ พร้อมทั้งจำลอง Contact Window ที่เปิดและปิดตามรอบเวลา และใช้ Priority Queue ในการจัดการข้อมูลแบบ Store-and-Forward
- พัฒนา server.py สำหรับสถานีภาคพื้นดินที่รับข้อมูลจากดาวเทียมผ่าน TCP วัดค่า End-to-End Latency และประมวลผลข้อมูลผ่านระบบหลายเธรด

ผลงานที่ได้คือระบบจำลองการสื่อสารยานอวกาศ-ดาวเทียมที่สามารถทนต่อความหน่วงสูงและสัญญาณขาดช่วงได้อย่างมีประสิทธิภาพ

สิ่งที่ได้เรียนรู้
- DTN(Delay/Disruption-Tolerant Networking): การสื่อสารแบบ Store-and-Forward ที่ทนต่อความหน่วงสูงและสัญญาณขาดช่วง

- Mission QoS: การจัดลำดับความสำคัญของข้อมูลตามผลกระทบต่อภารกิจ ไม่ใช่ตามขนาดไฟล์

- Contact Window: การจำลองช่วงเวลาที่ลิงก์สื่อสารเปิด/ปิด และการส่งข้อมูลให้สอดคล้องกับช่วงเวลานั้น

- Priority Queue & Preemption : การจัดการคิวที่ข้อมูลฉุกเฉินสามารถแทรกและส่งก่อนข้อมูลทั่วไปได้ทันที

- End-to-End Latency Measurement: การวัดเวลาตั้งแต่ข้อมูลถูกสร้างจนถึงปลายทาง
